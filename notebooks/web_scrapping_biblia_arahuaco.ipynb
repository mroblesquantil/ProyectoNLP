```python
data = {}
url='https://media.ipsapps.org/arh/osa/bible/01-GEN-001.html'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
i = True
sig_pag = "01-GEN-001.html"

while i == True:
    print(sig_pag)
    url='https://media.ipsapps.org/arh/osa/bible/' + sig_pag
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    content_div = soup.find('div', id='content')
    
    if content_div.get_text() != '\n\n\n\n\xa0\n':
        # Encontrar el div con id="Ts1"
        div_to_exclude = content_div.find('div', {'id': 'Ts1'})

        # Eliminar el div y su contenido del árbol
        div_to_exclude.extract()

        # Obtener el texto del contenido restante
        texto_final = content_div.get_text()

        data[sig_pag.split(".")[0]] = texto_final
    
    sig_pag = soup.find('a', {'title': 'Next Chapter'}).get('href')
    
    if sig_pag == "":
        i = False
    
    
```
